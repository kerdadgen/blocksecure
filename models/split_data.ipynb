```python
import pandas as pd
```

```python
df = pd.read_csv('data/transaction_dataset.csv')
df.head()
```

```python
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)

# SÃ©parer les 40 transactions pour le test
test_df = df_shuffled.iloc[:40]

# Le reste pour l'entraÃ®nement
train_df = df_shuffled.iloc[40:]
```

```python
test_df
```

```python

# Sauvegarder les deux jeux
test_df.to_csv('data/transactions_test.csv', index=False)
train_df.to_csv('data/transactions_train.csv', index=False)

print(f"- Fichier test : {len(test_df)} transactions â†’ transactions_test.csv")
print(f"- Fichier entraÃ®nement : {len(train_df)} transactions â†’ transactions_train.csv")
```
